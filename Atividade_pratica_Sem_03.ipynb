### Sobre o dataframe:

O conjunto de dados contém informações sobre bebês nascidos e seus pais.

O **dicionário de dados** da base é o seguinte:

* **ID**: número do bebê
* **Length**: comprimento do bebê (cm)
* **Birthweight**: peso do bebê (kg)
* **Headcirc**: circunferência da cabeça
* **Gestation**: gestação (semanas)
* **smoker**: mãe fumante (1 = sim, O= não)
* **mage**: idade da mãe
* **mnocig**: número de cigarros por dia da mãe
* **mheight**: altura da mãe (cm)
* **mppwt**: peso pré-gravidez da mãe (kg)
* **fage**: idade do pai
* **fedyrs**: anos de escolaridade do pai
* **fnocig**: número de cigarros por dia do pai
* **fheight**: altura do pa (cm)
* **lowbwt**: baixo peso do bebê (0 = não; 1 = sim)
* **mage35**: mãe acima dos 35 anos (0 = não; 1 = sim)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/nagelamartins/Mineracao_de_dados/refs/heads/main/Base-dados_Sem-03.csv'
df = pd.read_csv(url)
df.head(20)
```

```python
df.info()
```

Utilizando a função `info()` é possível entender que não existem dados ausentes a serem imputados.

```python
# Histograma do atributo peso do bebê.

sns.histplot(df['Birthweight'])
plt.title('Peso do bebê')
plt.xlabel('Peso (kg)')
plt.ylabel('Frequência absoluta')
plt.show()
```

```python
# Histograma do atributo altura do bebê.

sns.histplot(df['Length'])
plt.title('Altura do bebê')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência absoluta')
plt.show()
```

```python
# Gráfico de setores das mães fumantes.

mae_fumante = len(df[df['smoker'] == 1])
mae_nao_fumante = len(df[df['smoker'] == 0])

maes = [mae_fumante, mae_nao_fumante]
rotulos = ['Fumante', 'Não fumante']

plt.pie(maes, labels = rotulos, autopct = '%0.0f%%')
plt.title('Proporção de mães fumantes e não fumantes')
plt.show()
```

```python
# Média e o desvio-padrão da altura dos bebês.

df['Length'].describe()
```

```python
# Diagrama de caixa da idade da mãe.

sns.boxplot(y = df['mage'])
plt.title('Idade da mãe')
plt.ylabel('Idade')
plt.show()
```

```python
# Gráfico de dispersão do peso do bebê e do tempo de gestação.

sns.scatterplot(x = df['Birthweight'], y = df['Gestation'])
plt.title('Peso do bebe x Tempo de gestação')
plt.xlabel('Peso do bebe')
plt.ylabel('Tempo de gestação')
plt.show()
```

```python
# Informação de mães fumantes e não fumantes acrescentadas.

smoker_rotulos = {0: 'Não', 1: 'Sim'}
df['mae_fumante'] = df['smoker'].map(smoker_rotulos)

sns.scatterplot(x = df['Birthweight'], y = df['Gestation'], hue = df['mae_fumante'])
plt.title('Peso do bebe x Tempo de gestação para mãe fumantes e não fumantes')
plt.xlabel('Peso do bebe')
plt.ylabel('Tempo de gestação')
plt.show()
```
